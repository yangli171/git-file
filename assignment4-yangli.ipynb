```python
#Import necessary Bokeh modules
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, NumeralTickFormatter
from bokeh.io import output_notebook, show
from bokeh.palettes import RdBu
from bokeh.transform import linear_cmap, jitter
```

```python
#Make groupby into a new DataFrame
gender_df = gender_artist_df = artist_df.groupby('gender')

# Set up the source data that will suppply the x,y columns and the film title hover text
source = ColumnDataSource(gender_df)

# Set the hover tool tip to the film title, release year, and proportion of dialogue

TOOLTIPS = [("Name", "@name"),
            ("Year", "@yearOfBirth"),
           ("Place", "@placeOfBirth")]

#Set up Bokeh plot with title, labels, 
bokeh_plot = figure(title="Year of birth difference between male and female artists", x_axis_label = 'yearOfBirth_std',
                    y_axis_label = 'gender',x_range = [1543, 2014], y_range = ['Male','Female'],
                 tooltips=TOOLTIPS, width=800, height=550, active_scroll='wheel_zoom')

# Create a red to blue color palette
color_mapper = linear_cmap(field_name='yearOfBirth_std', palette=RdBu[4], low=1.1, high=0)

# Supply inidivudal points values
bokeh_plot.circle(y='yearOfBirth_std', x=jitter('gender', width=.2),
         size = 10,
        line_color='black',
        line_alpha=.4,
         source=source,
         color=color_mapper, alpha=.5)

bokeh_plot.title.text_font_size='20pt'

show(bokeh_plot)
```
